```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
nome_ativo = 'PETROBRAS-AMBEVSA-VALE-ITAUSA-JBS-BRASIL-3RPETROLEUM-AZUL-ASSAI-TOTVS'
ativos = ['PETR3.SA','ABEV3.SA', 'VALE3.SA','ITSA4.SA', 'JBSS3.SA','BBAS3.SA','RRRP3.SA','AZUL4.SA','ASAI3.SA','TOTS3.SA']
data_final = datetime.today() - timedelta(days=(30))
```

```python
dados1 =[]
for ativo in ativos:
    data_inicial = data_final - timedelta(days=(365+30))
    dados = yf.download(ativo, start=data_inicial, end=data_final)
    dados1.append(dados)
```

```python
dados5 =[]
for ativo in ativos:
    data_inicial = data_final - timedelta(days=((365*5)+30))
    dados = yf.download(ativo, start=data_inicial, end=data_final)
    dados5.append(dados)
```

```python
dados10 =[]
for ativo in ativos:
    data_inicial = data_final - timedelta(days=((365*10)+30))
    dados = yf.download(ativo, start=data_inicial, end=data_final)
    dados10.append(dados)
```

```python
dados20 =[]
for ativo in ativos:
    data_inicial = data_final - timedelta(days=((365*20)+30))
    dados = yf.download(ativo, start=data_inicial, end=data_final)
    dados20.append(dados)
```

```python
for dados in dados1:
    dados.reset_index(inplace=True)
    dados['mm9'] = 0
    for i in range(9, len(dados)):
        dados['mm9'].loc[i] = dados['Adj Close'].loc[i-9:i-1].mean()
    dados['mm14'] = 0
    for i in range(14, len(dados)):
        dados['mm14'].loc[i] = dados['Adj Close'].loc[i-14:i-1].mean()
    dados['mm21'] = 0
    for i in range(21, len(dados)):
            dados['mm21'].loc[i] = dados['Adj Close'].loc[i-21:i-1].mean()
    dados['vwap9'] = 0
    for i in range(9, len(dados)):
        soma, peso = 0,0
        for j in range(9):
            soma += dados['Adj Close'].loc[i-j-1]*dados['Volume'].loc[i-j-1]
            peso += dados['Volume'].loc[i-j-1]
        dados['vwap9'].loc[i] = soma/peso
    dados['vwap14'] = 0
    for i in range(14, len(dados)):
        soma, peso = 0,0
        for j in range(14):
            soma += dados['Adj Close'].loc[i-j-1]*dados['Volume'].loc[i-j-1]
            peso += dados['Volume'].loc[i-j-1]
        dados['vwap14'].loc[i] = soma/peso
    dados['vwap21'] = 0
    for i in range(21, len(dados)):
        soma, peso = 0,0
        for j in range(21):
            soma += dados['Adj Close'].loc[i-j-1]*dados['Volume'].loc[i-j-1]
            peso += dados['Volume'].loc[i-j-1]
        dados['vwap21'].loc[i] = soma/peso
    dados['linhak'] = 0
    for i in range(14, len(dados)):
        c = dados['Adj Close'].loc[i]
        Ln = dados['Low'].loc[i-14:i-1].min()
        Hn = dados['High'].loc[i-14:i-1].max()
        dados['linhak'].loc[i] = (c-Ln)/(Hn-Ln)*100
    dados['linhad'] = 0
    for i in range(17, len(dados)):
        ld = dados['linhak'].loc[i-3:i-1].mean()
        dados['linhad'].loc[i] = ld
    dados['rsi9'] = 0
    for i in range(9, len(dados)):
        dados['rsi9'].loc[i] = 100-100/(1+((dados['High'].loc[i-9:i-1].mean())/(dados['Low'].loc[i-9:i-1].mean())))
    dados['rsi14'] = 0
    for i in range(14, len(dados)):
        dados['rsi14'].loc[i] = 100-100/(1+((dados['High'].loc[i-14:i-1].mean())/(dados['Low'].loc[i-14:i-1].mean())))
    dados['rsi21'] = 0
    for i in range(21, len(dados)):
        dados['rsi21'].loc[i] = 100-100/(1+((dados['High'].loc[i-21:i-1].mean())/(dados['Low'].loc[i-21:i-1].mean())))
    dados['linhamacd'] = 0
    for i in range(26, len(dados)):
        mme12 = dados['Adj Close'].loc[i-26:i-15].mean()
        mme26 = dados['Adj Close'].loc[i-26:i-1].mean()
        linhamacd = mme12 - mme26
        dados['linhamacd'].loc[i] = linhamacd
    dados['sinalmacd'] = 0
    for i in range(9, len(dados)):
        mme9 = dados['linhamacd'].loc[i-9:i-1].mean()
        dados['sinalmacd'].loc[i] = mme9
    dados['bbsuperior'] = 0
    dados['bbinferior'] = 0
    dados['bbmeio'] = 0
    for i in range(20, len(dados)):
        lc = dados['Adj Close'].loc[i-20:i-1].mean()
        dp = np.std(dados['Adj Close'].loc[i-20:i-1])
        bs = lc+2*dp
        bi = lc-2*dp
        dados['bbmeio'].loc[i] = lc
        dados['bbsuperior'].loc[i] = bs
        dados['bbinferior'].loc[i] = bi
    dados['target'] = 0
    for i in range(1,len(dados)):
        if dados['Adj Close'].loc[i]>=dados['Adj Close'].loc[i-1]:
            dados['target'].loc[i] = 1
        else:
            dados['target'].loc[i] = 0
    dados = dados[29:]
```

```python
models = [LogisticRegression(),
          RandomForestClassifier(),
          SVC(),
          KNeighborsClassifier(),
          MLPClassifier()]

for m in models:
    for metrica in range(len(metricas)):
        model = m
        testeX=[]
        testeY=[]
        for d,dados in enumerate(dados1):
            X = metricas[metrica]
            Y = ['target']
            X = dados[X]
            Y = dados[Y]
            trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.3, random_state=42)
            testeX.append(testX)
            testeY.append(testY)
            model.fit(trainX,trainY)        
        testXgeral = pd.concat(testeX, ignore_index=True)
        testYgeral = pd.concat(testeY, ignore_index=True)
        previsao = model.predict(testXgeral)
        print(accuracy_score(testYgeral, previsao))
```
